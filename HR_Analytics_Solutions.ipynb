```python
train_data = pd.get_dummies(train_data,columns = ['relevent_experience','gender','enrolled_university','education_level','major_discipline','company_size','company_type','last_new_job'])
train_data = train_data.drop(['relevent_experience_No relevent experience','gender_Other','enrolled_university_missing','education_level_missing','major_discipline_missing','company_size_5000-9999','company_type_Other','last_new_job_missing'],axis = 1)
```

```python
train_data
```

```python
def func(t):
  if t == '>20':
    return 21
    
  elif t == '<1':
    return 0.5
    
  elif t == 'missing':
    return 0
    
  else:
    return t

train_data.experience = train_data.experience.apply(func )
```

```python
train_data['experience'].value_counts()
```

```python
train_data = train_data.drop(['city'],axis =1)
```

```python
X = train_data.drop(['target'],axis =1)
y = train_data['target']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.metrics import accuracy_score

model = LogisticRegression(class_weight='balanced')
model = model.fit(X_train, y_train)
yhat = model.predict(X_test)
accuracy = accuracy_score(y_test, yhat)
print('Accuracy: %.2f' % (accuracy*100))
```
